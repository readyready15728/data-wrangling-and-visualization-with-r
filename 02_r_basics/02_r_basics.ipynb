I'm going to do one (or two) better than the book and use a function to calculate the two roots, in addition to making the discriminant a complex number:

````r
quadratic_formula <- function(a, b, c) {
    discriminant <- sqrt(as.complex(b^2 - 4 * a * c))
    r_1 <- (-b - discriminant) / (2 * a)
    r_2 <- (-b + discriminant) / (2 * a)

    list(r_1, r_2)
}
````

### Factors

`factor`s are for categorical data and they look like character strings, but under the hood R assigns them to distinct integers in order to be more efficient. These coded categories can be seen with the `levels()` function:

````r
levels(murders$region)
````

I'm going to crib a good deal from the book here:

"Note that the levels have an order that is different from the order of appearance in the `factor` object. The default in R is for the levels to follow alphabetical order. However, often we want the levels to follow a different order. You can specify an order through the `levels` argument when creating the `factor` with the `factor()` function. For example, in the `murders` dataset regions are ordered from east to west. The function `reorder()` lets us change the order of the levels of a `factor` variable based on a summary computed on a `numeric` vector."

\[There are a number of ways to specify ordinal `factor`s but let's just look at this one for now.\]

"Suppose we want the levels of `region` ordered by the total number of murders rather than alphabetically. If there are values associated with each level, we can use the `reorder()` function and specify a data summary to determine the order. The following code takes the sum of the total murders in each region, and reorders the `factor` following these sums."

````r
region <- murders$region
value <- murders$total
region <- reorder(region, value, FUN = sum)
levels(region)
````

## Sorting

Sorting is the next feature of R to be described and the book motivates the material with an analysis of the `murders` dataset.

### `sort()`

"Say we want to rank the states from least to most gun murders. The function `sort()` sorts a vector in increasing order. We can therefore see the largest number of gun murders by typing:"

````r
sort(murders$total)
````

"However, this does not give us information about which states have which murder totals. For example, we don’t know which state had 1257."

### `order()`

`sort()` is not what we want, but `order()`, which takes a vector as input and returns a vector of indices that sorts the input vector. Here is a simplified example that shows how these two functions differ:

````r
x <- c(31, 4, 15, 92, 65)
index <- order(x)
index
````

Note that `x[index]` returns the same thing as `sort()` would have:

````r
x[index]
````

Why does this help us? Because `murders` is not just a vector, but an entire `data.frame`. Thus the result of `order()` can be used one other columns within the `data.frame`:

````r
index <- order(murders$total)
murders$abb[index]
````

### `max()` and `which.max()`

If only interested in the entry with the largest value, `max()` and `which.max()` (gives the index of the largest value) can be used:

````r
max(murders$total)
````

````r
murders$state[which.max(murders$total)]
````

(`min()` and `which.min()` work in a similar fashion.)

````r
min(murders$total)
````

````r
murders$state[which.min(murders$total)]
````

### `rank()`

There is also a less used function called `rank()` which gives the rank of each entry in a vector, starting with 1 for the lowest value and leading up to $n$ for the highest value. (There's also tie-breaking but I'm not concerned with that right now.) Example:

````r
x <- c(31, 4, 15, 92, 65)
rank(x)
````

### Two vectors

Those were operations with a scalar and vector; now consider two vectors. Here is what element-wise addition of two `numeric` or `integer` vectors, each of length 4 (just for example), looks like:

$$
\begin{pmatrix}
a \\
b \\
c \\
d
\end{pmatrix}
+
\begin{pmatrix}
e \\
f \\
g \\
h
\end{pmatrix}
=
\begin{pmatrix}
a + e \\
b + f \\
c + g \\
d + h
\end{pmatrix}
$$

Addition is one of a number of vectorized operations in R. `*` for example behaves in the same way. And that's how we can get *per capita* statistics in terms of homicides per 100,000 citizens for each of the states:

````r
murder_rate <- murders$total / murders$population * 100000
````

With `murder_rate`, it is now possible to order the states and DC in by homicide rate, ascending:

````r
murders$abb[order(murder_rate)]
````

### Beware of recycling

On to the concept of recycling: "\[A\] common source of unnoticed errors in R is the use of recycling. We saw that vectors are added elementwise. So if the vectors don't match in length, it is natural to assume that we should get an error. But we don’t. Notice what happens:"

````r
x <- c(1, 2, 3)
y <- c(10, 20, 30, 40, 50, 60, 70)
x + y
````

"We do get a warning, but no error. For the output, R has recycled the numbers in `x`. Notice the last digit of numbers in the output."

## Indexing

The `murders` dataset will now be used to motivate the topic of indexing in R.

### Subsetting with `logical`s

"Imagine you are moving from Italy where, according to an ABC News report, the murder rate is only 0.71 per 100,000. You would prefer to move to a state with a similar murder rate. Another powerful feature of R is that we can use logicals to index vectors. If we compare a vector to a single number, it actually performs the test for each entry. The following is an example related to the question above:"

````r
ind <- murder_rate < 0.71
````

Non-strict inequality is also an option of course:

````r
ind <- murder_rate <= 0.71
````

The result is a `logical` vector with `TRUE` for instances where the predicate at the index is true and `FALSE` where the predicate is false. The resulting `logical` vector can be used to subset both vectors and `data.frame`s. Here's one such example:

````r
murders$state[ind]
````

It's also possible to use `sum()` on the `logical` vector which will coerce the `logical` values to `integer`s with `TRUE` as 1 and `FALSE` as 0. Thus `sum()` provides a count of "hits":

````r
sum(ind)
````

### Logical operators

Now for operations on `logical` vectors. Suppose you want to move to a safe state in the Western United States, perhaps for mountainous terrain or temperate rainforests. "Safe" will be quantified as having a homicide rate less than or equal to 1 per 100,000. In this case, `&` can be used to carry out a Boolean *and* operation on two `logical` vectors:

````r
west <- murders$region == "West"
safe <- murder_rate <= 1
murders$state[west & safe]
````

Now consider looking up the homicide rate of a given state, specifically California. It is possible to use a `logical` vector testing equality:

````r
murder_rate[murders$state == 'California']
````

### `which()`

However, there is also a function called `which()` that outputs which indices of a `logical` vector are `TRUE` and can be used for indexing a matching vector or `data.frame`:

````r
ind <- which(murders$state == 'California')
ind
````

````r
murder_rate[ind]
````

### `match()`

Multiple indices can be returned with `match()`:

````r
ind <- match(c('New York', 'Florida', 'Texas'), murders$state)
murder_rate[ind]
````

### `%in%`

If, instead of indices we want a `logical` vector that tells whether or not each element of the first vector is a member of the second, we can use the `%in%` operator:

````r
c('Boston', 'Dakota', 'Washington') %in% murders$state
````

`match()` is nearly equivalent to combining `which()` with `%in%` (but notice the order of the following results):

````r
match(c('New York', 'Florida', 'Texas'), murders$state)
````

````r
which(murders$state %in% c('New York', 'Florida', 'Texas'))
````
